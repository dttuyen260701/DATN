```python
#dataset1
df1=pd.read_excel('streetname.xlsx')
df1
# df1.describe()
# df1.info()
```

Hàm lấy chiều rộng đường

```python
def getStreetWitdh(content):
  contentArray = content.split("\n")
  for item in contentArray:
    if(item.find("Chiều rộng") != -1):
      return item.replace("2. Chiều rộng", "").replace(":", "")
  return "0"
```

Lấy tên đường và chiều rộng sau đó đổi tên cột

```python
dataStreet = df1[['Mô tả', 'Tên đường']]
dataStreet['Tên đường'] = dataStreet['Tên đường'].apply(lambda x: "Phan Đình Giót" if(x == "Liên Chiểu") else x)
dataStreet['Mô tả'] = dataStreet['Mô tả'].apply(lambda x: getStreetWitdh(x) if(x != "" and type(x) == str) else "0")
dataStreet["STT"] = dataStreet.index + 1
dataStreet.rename(columns={'Mô tả':'width'}, inplace=True)
dataStreet.rename(columns={'Tên đường':'name'}, inplace=True)
dataStreet
```
