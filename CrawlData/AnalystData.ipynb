Hàm xem các định dạng sau số

```python
def findAllFormat(data, partern):
  last_rs = []
  temp = list(data)
  for i in temp:
    rs = results_cc = re.findall(partern, str(i))
    for k in rs:
      last_rs.append(k)
  return set(last_rs)
findAllFormat(dataAloNhaDat['Width'], "(\D+)")
```

Đưa Column Width về chung đơn vị (m) và kiểu dữ liệu Float

```python
def convertWidth(data):
  data = data.replace(",",".").replace("..",".").replace("/", ".").replace("`","").replace(".m", "m").replace("m", "")
  if(len(data.split(".")) > 2): data = data[0:data.rfind('.')]
  return float(data)
findAllFormat(dataAloNhaDat['Width'], "(\D+)")
#tìm các giá trị biệt lệ và thay bằng replace như: /, `, .m -> về cùng 1 đơn vị là m -> convert về float
dataAloNhaDat['Width'] = dataAloNhaDat['Width'].apply(lambda x: convertWidth(x) if x != nan and type(x) == str else x)
dataAloNhaDat.info()
```

Hàm convert to CSV

```python
def createCSV(name, listToCSV):
  csv_columns = []
  for product in listToCSV:
      tempkhac = set(list(product)[0:]) ^ set(csv_columns)
      tempgiong = set(list(product)[0:]) & set(csv_columns)
      csv_columns = tempkhac ^ tempgiong
  csv_file = name
  try:
      with open(csv_file, 'w+',encoding= "utf-8") as csvfile:
          writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
          writer.writeheader()
          for dataz in listToCSV:
              try:
                  writer.writerow(dataz)
              except:
                  continue
  except IOError:
      print("I/O error")
```

Đưa Column Price về chung đơn vị triệu/m2 kiểu dữ liệu float (ID theo Enum)

```python
# Tìm tất cả định dạng đơn vị 
findAllFormat(dataAloNhaDat['Price'], "(\D+)")
# Result: {'\n  ', ' ngàn', ' triệu', ' triệu / m\n', ' tỷ', ' tỷ ', ',', 'Thỏa thuận'}
def convertMPerMetter(id):
  dataTemp = dataAloNhaDat[dataAloNhaDat['PricePerMet'] == id]
  value = dataTemp.iloc[0]['Price']
  square = dataTemp.iloc[0]['Square']
  pattern_mPerMetter = "(\d+|\d+.\d+)(?= triệu / m)"
  results_mPerMetter = re.findall(pattern_mPerMetter, value)
  pattern_Billion = "(\d+|\d+.\d+)(?= tỷ)"
  results_Billion = re.findall(pattern_Billion, value)
  pattern_Million = "(\d+|\d+.\d+)(?= triệu)"
  results_Million = re.findall(pattern_Million, value)
  if len(results_mPerMetter) > 0:
    return round(float(results_mPerMetter[0]), 3)
  if len(results_Billion) > 0:
    priceAll = round(float(results_Billion[0]), 3)
    pricePercen = round((priceAll*1000/square), 2)
    return pricePercen
  if len(results_Million) > 0:
    return round((float(results_Million[0])/square), 2)
  return None

dataAloNhaDat['PricePerMet'] = dataAloNhaDat.index + 1
dataAloNhaDat['Price'] = dataAloNhaDat['Price'].apply(lambda x: x.replace("\n", "").replace(",", ".") if x != nan and type(x) == str else nan)
dataAloNhaDat['PricePerMet'] = dataAloNhaDat['PricePerMet'].apply(lambda x: convertMPerMetter(x) if x != nan else x)
dataAloNhaDat = dataAloNhaDat.drop(columns=['Price'])

dataAloNhaDat.info()
dataAloNhaDat.describe()
```

```python
def RemoveLastParams(address):
  value = address.split(', ')
  return address.replace(', ' + value[len(value) - 1], "")
```

Thêm cột WardPoint tham chiếu với dataWardDistrict

```python
def getWardPoint(address, source):
  value = ""
  if(address.find('Phường') != -1):
    value = address.replace(address.split('Phường')[0], "")
  elif(address.find('Xã') != -1):
    value = address.replace(address.split('Xã')[0], "")
  else:
    return None
  resultFind = source[source['Name'] == value]
  if(len(resultFind) > 0):
    return resultFind.iloc[0]['Point']
  else:
    max_value = source['Point'].max()
    global dataWardDistrict 
    dataWardDistrict = source.append({'Name': value, 'Point': max_value + 1},ignore_index=True)
    return max_value + 1

dataAloNhaDat['WardPoint'] = dataAloNhaDat['Address']
dataAloNhaDat['WardPoint'] = dataAloNhaDat['WardPoint'].apply(lambda x: getWardPoint(x, dataWardDistrict) if x != nan else x)
#Xóa điểm cuối cùng của dữ liệu
dataAloNhaDat['Address'] = dataAloNhaDat['Address'].apply(lambda x: RemoveLastParams(x) if x != nan else x)
dataAloNhaDat.info()
dataAloNhaDat.describe()
```

Gán tên đường bằng Point

```python
def getRoadPoint(address, source):
  value = ""
  if(address.find('Đường') != -1):
    value = address.replace(address.split('Đường')[0], "").replace('Đường ', '').replace('.', '')
    if(value.find('Phường') != -1):
      value = value.replace(value.split(' Phường')[1], '').replace(' Phường', '')
  else:
    return None
  resultFind = source[source['Name'].str.lower() == value.lower()]
  if(len(resultFind) > 0):
    return resultFind.iloc[0]['Point']
  else:
    max_value = source['Point'].max()
    global dataStreet 
    dataStreet = source.append({'Name': value, 'Point': max_value + 1},ignore_index=True)
    return max_value + 1

def handleExceptionRoad(x):
    value = x.replace("Chế Lan Viên Chế Lan Viên", "Chế Lan Viên").replace("Đường 7,5m ", "").replace("số 1 nam kì khởi nghĩa", "Nam Kỳ Khởi Nghĩa")
    return value.replace("Bùi Tá Hán Bùi Tá Hán", "Bùi Tá Hán").replace("	Trường Thi 4 Nguyễn Văn Linh", "	Trường Thi 4").replace("Đường Bạch Đằng", "Đường Bạch Đằng")

dataAloNhaDat['RoadPoint'] = dataAloNhaDat['Address'].apply(lambda x: handleExceptionRoad(x) if x != nan else x)
dataAloNhaDat['RoadPoint'] = dataAloNhaDat['RoadPoint'].apply(lambda x: getRoadPoint(x, dataStreet) if x != nan else x)
dataAloNhaDat.info()
dataAloNhaDat.describe()
```

Get các value width từ dataStreet

```python
def getStreetWidth(id):
  result = nan
  dataTemp = dataAloNhaDat[dataAloNhaDat['WidthRoad'] == id]
  roadPoint = dataTemp.iloc[0]['RoadPoint']
  resultFind = dataStreet[dataStreet['Point'] == roadPoint]
  if len(resultFind) > 0:
    result = resultFind.iloc[0]['width']
    if not isnan(result):
      return result
    else:
      result = dataTemp.iloc[0]['RoadOfFront']
  else:
    result = dataTemp.iloc[0]['RoadOfFront']
  return result

dataAloNhaDat['WidthRoad'] = dataAloNhaDat.index + 1
dataAloNhaDat['WidthRoad'] = dataAloNhaDat['WidthRoad'].apply(lambda x: getStreetWidth(x))
dataAloNhaDat = dataAloNhaDat.drop(columns=['RoadOfFront'])
dataAloNhaDat.info()
dataAloNhaDat.describe()
```

Đưa Column RoadOfFront về chung đơn vị m kiểu dữ liệu float

```python
findAllFormat(dataAloNhaDat['WidthRoad'], "(\D+)")
# Result: {' ', ',', ',m', ',mm', '--m', '-m', ':', 'M', 'Mm', '`m', 'm', 'mm', 'nan'}

def converToMetter(value):
  value = value.replace('-', '').replace('M', '').replace(':', '').replace('mm', 'm').replace('`', '').replace(',', '.')
  pattern_Metter = "(\d+|\d+.\d+)(?=m)"
  result_Metter = re.findall(pattern_Metter, value)
  result = 0
  if len(result_Metter) > 1:
    result = round(float(result_Metter[0].replace(' ', '')) + float(result_Metter[1])/10, 3)
  if len(result_Metter) > 0:
    result = round(float(result_Metter[0].replace(' ', '')), 3)
  while(result > 56):
    result = round(result / 10, 3)
  return result if result > 0 else None

dataAloNhaDat['WidthRoad'] = dataAloNhaDat['WidthRoad'].apply(lambda x: converToMetter(x) if type(x) == str else x)
dataAloNhaDat = dataAloNhaDat.drop(columns=['Address'])
dataAloNhaDat.info()
dataAloNhaDat.describe()
```
